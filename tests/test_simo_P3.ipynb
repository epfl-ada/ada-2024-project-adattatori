```python
from geopy.geocoders import Nominatim
import pandas as pd

# Example DataFrame
df = df_institutions_count['Institution']

geolocator = Nominatim(user_agent="geoapi")
def get_location(name):
    location = geolocator.geocode(name)
    if location:
        return location.address
    return None

df_institutions_count['Location'] = df_institutions_count['Institution'].apply(get_location)
print(df_institutions_count)
```

```python
# Function to extract country
def extract_country(address):
    if address:
        # Split the address by commas and return the last element (which is often the country)
        return address.split(",")[-1].strip()
    return None

# Apply the function to extract countries
df_institutions_count['Country'] = df_institutions_count['Location'].apply(extract_country)

# Display the updated DataFrame
df_institutions_count['Country'].value_counts()
```

```python
from geopy.geocoders import Nominatim
import pandas as pd


# Initialize the geolocator
geolocator = Nominatim(user_agent="geoapi")

# Function to get address in English
def get_location_in_english(name):
    try:
        location = geolocator.geocode(name, language='en')  # Request results in English
        if location:
            return location.address
        return 'Address not found'
    except Exception as e:
        print(f"Error for {name}: {e}")
        return 'Error'

# Apply the function to get addresses
df_institutions_count['Location'] = df_institutions_count['Institution'].apply(get_location_in_english)

# Extract country
def extract_country(address):
    if address and address != 'Address not found' and address != 'Error':
        return address.split(",")[-1].strip()
    return None

df_institutions_count['Country'] = df_institutions_count['Location'].apply(extract_country)

print(df_institutions_count)
```

```python
import geodatasets
# Load the world map dataset

url = "https://naciscdn.org/naturalearth/110m/cultural/ne_110m_admin_0_countries.zip"

world = gpd.read_file(url)
# Export to GeoJSON to simplify invalid geometries
world.to_file("world_fixed.geojson", driver="GeoJSON")

# Reload the fixed GeoJSON
world = gpd.read_file("world_fixed.geojson")

for i, col in enumerate(world.columns):
    print(f'{i}: {col}')
```

```python
data.head()
```

```python
# Merge the world map with the DataFrame
world = world.merge(data, how='left', left_on='SOVEREIGNT', right_on='Country')
world.head()
```

```python


# Replace NaN counts with 0
world['count'] = world['count'].fillna(0)

# Plot the map
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.boundary.plot(ax=ax, linewidth=1, color="black")
world.plot(column='count', ax=ax, legend=True, cmap='viridis',
           legend_kwds={'label': "Number of Institutions by Country", 'orientation': "horizontal"})

# Add title and remove axes
plt.title('Global Distribution of Research Institutions', fontsize=16)
plt.axis('off')
plt.show()
```
